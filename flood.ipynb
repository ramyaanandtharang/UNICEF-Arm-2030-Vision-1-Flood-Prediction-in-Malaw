```python
def cap_data(flood):
    for col in flood.columns:
        print("capping the ",col)
        if (((flood[col].dtype)=='float64') | ((flood[col].dtype)=='int64')):
            percentiles = flood[col].quantile([0.25,0.75]).values
            flood[col][flood[col] <= percentiles[0]] = percentiles[0]
            flood[col][flood[col] >= percentiles[1]] = percentiles[1]
        else:
            flood[col]=flood[col]
    return flood

flood1=cap_data(flood)
```
